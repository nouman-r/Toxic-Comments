## Gensim

Gensim for Topic MOdelling using LDA

1. First need to Tokenize the text. 
    using gensim.utils.simple_preprocess()
2. Create Bigrams
3. Clean the Tokens (remove stop words, create bigrams and lemmatize)
4. Create dictionary of the cleaned tokens, using Gensim Dictionary. 
5. Create the corpus (encoded tokens with frequency distribution of each token) using  dictionary. 
6. Use the Corpus, Dictionary and number of topic required, create topics using LDA. 
7. Plot using pyLDAvis package.

```python
ldamodel = LdaModel(corpus=corpus, num_topics=15, id2word=d)
```
